```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
df['Body']=df['Body'].apply(remove_stopwords)
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
for i in range(len(df['Body'])):
    en=[]
    org=[]
    ct=[]
    dt=[]
    if (pd.isnull(df['Body'][i]) == False):
        df['Body'][i]=nlp(df['Body'][i])
        for j in df['Body'][i].ents:
            if(j.label_=='PERSON'):
                en.append(j.text)
            if(j.label_=='DATE'):
                dt.append(j.text)
            if(j.label_=='GPE'):
                ct.append(j.text)
            if(j.label_=='ORG'):
                org.append(j.text)

    df['Names'][i]=en
    df['Dates'][i]=dt
    df['Citys'][i]=ct
    df['Organizations'][i]=org
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('distilbert-base-nli-mean-tokens')
```

```python
content=df['Body'].to_list()
```

```python
embeddings=model.encode(content)
```

```python
import faiss
import numpy as np
```

```python
index=faiss.IndexIDMap(faiss.IndexFlatIP(768))
index.add_with_ids(embeddings,np.array(range(0,len(content))))
```

```python
faiss.write_index(index,'content_idx')
index=faiss.read_index('content_idx')
```

```python
def search(q):
    qe=model.encode([q])
    num=index.search(qe,2)
    return [content[_id] for _id in num[1].tolist()[0]]
```

```python
for re in search('We encrypted the network of this company'):
    print('\t',re)
```
